```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

Reading the Image 

```python
img = cv2.imread("/content/Ben 10 image.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(img)
plt.show()
```

Converting into Grayscale Image 

```python
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.axis('off')
plt.imshow(gray_img)
plt.show()
```

Invertion of The Gray Scale Image 

```python
#inv_gray_img = 255 - gray_img
inv_gray_img = cv2.bitwise_not(gray_img)
plt.axis('off')
plt.imshow(inv_gray_img)
plt.show()
```

Blurring the image 

```python
blur_img = cv2.GaussianBlur(inv_gray_img,(21,21),0)
plt.axis('off')
plt.imshow(blur_img)
plt.show()
```

Inverted Blur Image 

```python
#inv_blur_img = 255 - blur_img
inv_blur_img = cv2.bitwise_not(blur_img)
plt.axis('off')
plt.imshow(inv_blur_img)
plt.show()
```

```python
#Saving the image 
cv2.imwrite("./output_image.png",img)
```

Finally Make pencil Sketch of the image 

```python
ben_img = cv2.divide(gray_img,inv_blur_img,scale = 267.0)
plt.axis('off')
plt.imshow(ben_img)
plt.show()
```
